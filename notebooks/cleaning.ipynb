```python
import pandas as pd
file_path = r"C:\Work\DEPI graduation project\data\processed_movies.csv"

# Read the CSV into a DataFrame
df = pd.read_csv(file_path)
```

```python
df.info()
```

```python
# drop by name, return a new DataFrame
df = df.drop(columns=["adult", "imdb_id"])
df.info()
```

```python
df = df.drop(columns=["overview_filled", "id_overview"])
```

```python
df.info()
```

```python
df = df.dropna(subset=['overview', 'genres'])
```

```python
df.info()
```

```python
def normalize_list_field(value):
    if isinstance(value, list):
        return ", ".join(value)
    if pd.isna(value):
        return ""
    return str(value)

df["genres_text"] = df["genres"].apply(normalize_list_field)
df["companies_text"] = df["production_companies"].apply(normalize_list_field)

df["id_overview"] = (
    "ID: " + df["id"].astype(str)
    + " | Overview: " + df["overview"].fillna("")
    + " | Genres: " + df["genres_text"]
    + " | Production companies: " + df["companies_text"]
)
```
